```python
counts = df['Weather_Condition'].value_counts()
plt.pie(counts.values, labels=counts.index, startangle=140)
plt.title('Distribution of Weather Conditions')
plt.axis('equal')
plt.show() 
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x= 'Country', data=df, palette='pastel', hue='Country', legend=False )
plt.title('Distribution Of Accidents By Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show() 
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Severity', data=df, palette='pastel', hue='Severity', legend=False)
plt.title('Distribution Of Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.show()
```
